```python
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv(r'datasets\carSales.csv')
df.columns
df = df.drop(columns=['Customer Name','Dealer_Name','Phone','Company','Dealer_No ','Car_id','Date','Dealer_Region'])
df.head(5)
```

```python
isnacount = df.isna().sum()
isnullcount = df.isnull().sum()
print("NA data counts: \n", isnacount)
print("\nNULL data counts: \n", isnullcount)
```
